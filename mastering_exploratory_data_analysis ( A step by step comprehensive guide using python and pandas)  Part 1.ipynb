### 4. Handling Missing Data


```python
# Check for missing values
print("Missing values in each column:")
data.isnull().sum()
```

```python
# convert the 'LEVY' column to a numeric type and treat '-' as missing, 
data['LEVY'] = data['LEVY'].replace('-', None).astype(float)
data
```

```python
# Drop rows with missing values
print("Data after dropping missing values:")
data_drop = data.dropna()
data_drop
```

```python
# # Alternatively, fill missing values with a specified value for categorical feature or numerical features
# data_filled = data.fillna('NA')

# # Forward fill fills the missing values with the most recent non-missing value found before them in the column.( Mostly used in time series data ) Let's skip it for now since our data is not time series
# data['LEVY'].fillna(method='ffill', inplace=True)

# # Interpolate missing values (useful for time series data )
# data['LEVY'] = df['LEVY'].interpolate()
# print("Data after filling missing values with mean and interpolation:")
# data

# Imputing missing values with the mean of the column
data['LEVY'].fillna(data['LEVY'].mean(), inplace=True)
print("Data after filling missing values with mean: ")
data
```

### 10. Filtering Data


```python
# filter the DataFrame for 'TOYOTA' cars with a price greater than 10,000
filtered_data = data[(data['MANUFACTURER'] == 'TOYOTA') & (data['PRICE'] > 10000)]

# isplay the filtered DataFrame
print("Filtered DataFrame (TOYOTA cars with PRICE > 10,000):")
filtered_data.head(7)
```

### 11. Sorting Data


```python
# Sort the DataFrame by 'PRICE' in descending order
sorted_data = data.sort_values(by='PRICE', ascending=False)

# Display the sorted DataFrame
print("The DataFrame sorted by 'PRICE' in descending order:")
sorted_data.head(7)
```

### 12. Removing Duplicates

```python
# Remove duplicate rows
print("The data shape before dropping duplicates :")
print ( data.shape)


data_no_duplicates = data.drop_duplicates()

# Print the DataFrame after removing duplicates
print("The data shape after dropping duplicates :")
print ( data_no_duplicates.shape)
data_no_duplicates
```

### 13. Resetting Index


```python
# reset the index with drop=False
# print("DataFrame resetting index with drop=False:")

# data_reset = data_no_duplicates.reset_index(drop=False)

print("DataFrame resetting index with drop=True:")

# reset the index with drop=True
data_reset = data_no_duplicates.reset_index(drop=True)

# Display the DataFrame after resetting the index
print("DataFrame after resetting index:")
data_reset
```

### 14. Grouping, Aggregating, and Renaming Columns


```python
# Group by 'MANUFACTURER' and calculate aggregate statistics
grouped_data = data_reset.groupby('MANUFACTURER').agg({
    'PRICE': 'mean',
    'AIRBAGS': 'max',
    'MODEL': 'count'
})


grouped_data
# # Rename the columns for clarity
grouped_data.rename(columns={'PRICE': 'AVERAGE_PRICE', 'AIRBAGS': 'MOST_COMMON_AIRBAGS', 'MODEL': 'CAR_COUNT'}, inplace=True)

# # Print the grouped and aggregated DataFrame
print("Grouped DataFrame with mean PRICE, max AIRBAGS, and count of cars for each MANUFACTURER:")
grouped_data
```

### 15. Adding New Columns Using numpy.where


```python


# Define a threshold for 'PRICE' to categorize cars
price_threshold = 15000

# Add a new column 'PRICE_CATEGORY' using np.where
data_reset['PRICE_CATEGORY'] = np.where(data_reset['PRICE'] > price_threshold, 'Expensive', 'Affordable')

# Print the DataFrame with the new column
print("DataFrame with the new 'PRICE_CATEGORY' column using np.where:")
data_reset.head()
```

### 16. Using ‘replace ‘ to Clean and Transforming Column Data


```python

# Step 1: Remove 'km' from 'MILEAGE' values using replace
data_reset['MILEAGE'] = data_reset['MILEAGE'].replace(' km', '', regex=True)

# # Step 2: Convert the 'MILEAGE' column to integer type
data_reset['MILEAGE'] = data_reset['MILEAGE'].astype(int)

# # Step 3: Rename the column to 'MILEAGE_KM'
data_reset= data_reset.rename(columns={'MILEAGE': 'MILEAGE_KM'})

# Print the DataFrame after cleaning and transforming
print("DataFrame after replacing and transforming 'MILEAGE' column:")
data_reset

# applying mulitple replace function simultaneously
# data_reset['DOORS'] = data_reset['DOORS'].str.replace('04-May', '4-5').str.replace('02-Mar', '2-4')
data_reset
```

### 17. Apply Function with lambda to Transform Data


```python
# Step 1: define the mapping dictionary
doors_mapping = {
    '04-May': '4-5',
    '02-Mar': '2-4'
}

# Step 2: Apply the transformation using lambda
data_reset['DOORS'] = data_reset['DOORS'].apply(lambda x: doors_mapping.get(x, x))

# Print the DataFrame after transforming the 'DOORS' column
print("DataFrame after transforming 'DOORS' column using apply with lambda:")
data_reset
```

### 18. Understanding Pivot Tables in Pandas


```python
# Create a pivot table with 'MANUFACTURER' and 'CATEGORY' as indices and 'PRICE' as values
# data_reset['PRICE'] = data_reset['PRICE'].round(0).astype(int)

pivot_table = pd.pivot_table(data_reset, values='PRICE', index='MANUFACTURER', columns='CATEGORY', aggfunc='mean')

# Print the pivot table
print("Pivot table showing the average 'PRICE' for each 'MANUFACTURER' and 'CATEGORY':")
pivot_table
```

### 19. Target Data Transformation Techniques


```python
# Calculate cumulative sum for 'PRICE'
data_reset['CUMULATIVE_SUM_PRICE'] = data_reset['PRICE'].cumsum()

# Calculate percent change for 'PRICE'
data_reset['PERCENT_CHANGE_PRICE'] = data_reset['PRICE'].pct_change()

# Calculate expanding sum for 'PRICE'
data_reset['EXPANDING_SUM_PRICE'] = data_reset['PRICE'].expanding().sum()

# Calculate rank for 'PRICE'
data_reset['PRICE_RANK'] = data_reset['PRICE'].rank()

# Calculate difference for 'PRICE'
data_reset['PRICE_DIFF'] = data_reset['PRICE'].diff()

# Calculate log transformation for 'PRICE'
data_reset['LOG_PRICE'] = np.log(data_reset['PRICE'])

# Print the DataFrame after applying transformations
print("DataFrame after applying various transformations:")
data_reset[['PRICE','CUMULATIVE_SUM_PRICE','PERCENT_CHANGE_PRICE','EXPANDING_SUM_PRICE','PRICE_RANK','PRICE_DIFF','LOG_PRICE']]
```
